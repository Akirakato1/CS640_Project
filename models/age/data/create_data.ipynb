```python
'''
load and merge tweets and labels
'''
import pandas as pd
import numpy as np

labeled_users=pd.read_csv("./labeled_users.csv",lineterminator='\n')[['screen_name','age']]
cleaned_tweets=pd.read_excel("./cleaned_tweets.xlsx",engine='openpyxl').dropna(subset=['clean_text'])
data=pd.merge(cleaned_tweets, labeled_users, left_on='user_id', right_on='screen_name')
df=data[['screen_name','clean_text','age']]
```

```python
'''
    generate k-fold data of per tweet
'''
tdf=pd.DataFrame({'name': df['screen_name'], 'tweets': df['clean_text'],  'label': df['age']})
tdf.to_csv('./data/tweetswithlabel.csv', index=False)

# for k-fold data
name,indices=np.unique(df["screen_name"],return_index=True)
sorted=np.argsort(indices)
indices=indices[sorted]
name=name[sorted]


K=5
test_size=int(len(name)/K)
train_size=len(name)-test_size

for k in range(K):
    dfTest=tdf[tdf['name'].isin(name[k*test_size:(k+1)*test_size])]
    dfTrain=tdf[~tdf['name'].isin(name[k*test_size:(k+1)*test_size])]
    dfTest.to_csv('./data/testtweetswithlabel_{}.csv'.format(k+1), index=False)
    dfTrain.to_csv('./data/traintweetswithlabel_{}.csv'.format(k+1), index=False)

```

```python
'''
    merge tweets by usernames
'''
name,indices=np.unique(df["screen_name"],return_index=True)
sorted=np.argsort(indices)
indices=indices[sorted]
name=name[sorted]
data=[]
ndata=[]
label=[]

print(indices.shape)
# print(name)
print(name.shape)
for i in range(len(name)-1):
    s=""
    print(str(i)+": "+ name[i]+" "+str(indices[i])+ " from {} to {}".format(indices[i],indices[i+1]))
    for j in range(indices[i],indices[i+1]):
        
        s=s+df["clean_text"][j]+" "
    data.append(s)
    ndata.append(name[i]+" "+s)
    label.append(df["age"][indices[i]])

s=""
for k in range(indices[-1],len(df["clean_text"])):
    s=s+df["clean_text"][k]
data.append(s)
ndata.append(name[-1]+" "+s)
label.append(df["age"][indices[-1]])
print("{}: {} from {} to {}".format(len(data)-1,name[-1],indices[-1],len(df["screen_name"])-1))

import csv
newdf=pd.DataFrame({'name': name, 'tweets': data, 'name_tweets': ndata, 'label': label})
newdf.to_csv('./data/mergetweetswithlabel.csv', index=False)
```
